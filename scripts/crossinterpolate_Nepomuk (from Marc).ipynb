```julia
using NPZ
using PyPlot
using LaTeXStrings
using QuanticsTCI
import TensorCrossInterpolation as TCI

rcParams = PyPlot.PyDict(PyPlot.matplotlib."rcParams")
rcParams["font.size"] = 9
rcParams["text.usetex"] = true
rcParams["font.family"] = "serif"
rcParams["text.latex.preamble"] = [raw"\usepackage{amsmath}", raw"\usepackage{amssymb}"]
rcParams["font.serif"] = ["Computer Modern Roman"]
rcParams["lines.linewidth"] = 1.0
rcParams["lines.markersize"] = 2
rcParams["image.cmap"] = "coolwarm"
```

```julia
window = 500-256:500+255

freal = npzread("data/phs/T=0.01/Keldysh/NRG/multipoint_vertex_new/estep%3D16/hwidth%3D0.6/u%3D1.00/Asym_same_1222_imag.npy")
fimag = npzread("data/phs/T=0.01/Keldysh/NRG/multipoint_vertex_new/estep%3D16/hwidth%3D0.6/u%3D1.00/Asym_same_1222_real.npy")
f = freal[window, window] + 1im * fimag[window, window]

realv = max(maximum(abs.(freal)), maximum(abs.(fimag)))
kwords = Dict(:vmin=>-realv, :vmax=>realv)
```

```julia
fig, axs = subplots(ncols=2)
axs[1].imshow(real(f); kwords...) |> colorbar
axs[2].imshow(imag(f); kwords...) |> colorbar
```

```julia
qtt, ranks, errors = quanticscrossinterpolate(
    ComplexF64,
    (i, j) -> f[i, j],
    collect(collect.(axes(f))),
    tolerance=1e-5
)
```

```julia
semilogy(qtt.tt.pivoterrors)
```

```julia
semilogy(TCI.linkdims(qtt.tt))
```

```julia
qttf = qtt.(1:512, collect(1:512)')
```

```julia
fig, axs = subplots(ncols=2)
axs[1].imshow(real(qttf); kwords...) |> colorbar
axs[2].imshow(imag(qttf); kwords...) |> colorbar
```

```julia
fig, axs = subplots(ncols=2)
axs[1].imshow(real(f .- qttf)) |> colorbar
axs[2].imshow(imag(f .- qttf)) |> colorbar
```

```julia
filename = "/home/Anxiang.Ge/Desktop/PhD/mfrg_Data/Keldysh/1lpaper_data/parquetInit4_U_over_Delta=1.570796_T=0.010000_eVg=0.000000_n1=401_n2=201_n3=51_version1_final.h5"
fparquet = h5open(filename);
```

```julia
parquetK3t = map(t -> t.re + 1im * t.im, fparquet["K3_t"][1, 1, :, :, :, 1, 1])
parquetK3a = map(t -> t.re + 1im * t.im, fparquet["K3_a"][1, 1, :, :, :, 1, 1])
parquetK3p = map(t -> t.re + 1im * t.im, fparquet["K3_p"][1, 1, :, :, :, 1, 1])

parquetK2t = map(t -> t.re + 1im * t.im, fparquet["K2_t"][1, 1, :, :, 1, 1])
parquetK2a = map(t -> t.re + 1im * t.im, fparquet["K2_a"][1, 1, :, :, 1, 1])
parquetK2p = map(t -> t.re + 1im * t.im, fparquet["K2_p"][1, 1, :, :, 1, 1])

parquetK1t = map(t -> t.re + 1im * t.im, fparquet["K1_t"][1, 1, :, 1, 1])
parquetK1a = map(t -> t.re + 1im * t.im, fparquet["K1_a"][1, 1, :, 1, 1])
parquetK1p = map(t -> t.re + 1im * t.im, fparquet["K1_p"][1, 1, :, 1, 1])
```

```julia
using Interpolations

#Omegas = range(-1, 1; length=203)[2:end-1]
#omegas = @. Omegas * abs(Omegas) / sqrt(1 - Omegas ^2)
#omeganorm = fparquet["freq_params"]["bfreqs2_a"]["w_upper"][1, 1] / maximum(omegas)
#omegas *= omeganorm

#Nus = range(-1, 1; length=203)[2:end-1]
#nus = @. Nus * abs(Nus) / sqrt(1 - Nus^2)
#nunorm = fparquet["freq_params"]["ffreqs2_a"]["w_upper"][1, 1] / maximum(nus)
#nus *= nunorm
omegas = fparquet["bfreqs2_a"][:,1]
nus = fparquet["ffreqs2_a"][:,1]

K2ainterp = interpolate((omegas, nus), parquetK2a, Gridded(Linear()))
```

```julia
extent = 100
plotomegas = range(-extent, extent, length=300)
plotnus = range(-extent, extent, length=300)
interpdata = K2ainterp.(plotomegas, plotnus')

vmax = maximum(abs.(parquetK2a))
kwargs = Dict(:vmax=>vmax, :vmin=>-vmax)

fig, axs = subplots(ncols=2)
axs[1].imshow(real(interpdata); kwargs...) |> colorbar
axs[2].imshow(imag(interpdata); kwargs...) |> colorbar
```

```julia
R = 10
qR = 2^R
qmesh = range(-extent, extent, length=qR) |> collect

qttK2a, _, _ = quanticscrossinterpolate(
    ComplexF64,
    (x, y) -> K2ainterp(x, y),
    [qmesh, qmesh],
    tolerance=1e-2
)
qttK2a2, _, _ = quanticscrossinterpolate(
    ComplexF64,
    (x, y) -> K2ainterp(x, y),
    [qmesh, qmesh],
    tolerance=1e-3
)
```

```julia
qplotmesh = Int.(round.(range(1, qR, length=300)))
box = (-extent, extent, -extent, extent)

qttK2adata = qttK2a.(qplotmesh, qplotmesh')
fig, axs = subplots(ncols=2, nrows=2, figsize=(300, 250)./72)

axs[1, 1].set_title(L"\mathrm{Re}(K_{2a})")
axs[1, 1].imshow(real(qttK2adata)'; extent=box, kwargs...) |> colorbar
axs[1, 2].set_title(L"\mathrm{Im}(K_{2a})")
axs[1, 2].imshow(imag(qttK2adata)'; extent=box, kwargs...) |> colorbar

axs[1, 1].set_ylabel(L"\nu")
for ax in axs[1, :]
    ax.set_xlabel(L"\omega")
end

axs[2, 1].semilogy(1:TCI.rank(qttK2a.tt), qttK2a.tt.pivoterrors, label=L"\epsilon=10^{-2}")
axs[2, 1].semilogy(1:TCI.rank(qttK2a2.tt), qttK2a2.tt.pivoterrors, label=L"\epsilon=10^{-3}")
axs[2, 1].set_xlabel(L"D_{\max}")
axs[2, 1].set_ylabel("abs. error")

axs[2, 2].semilogy(1:2R-1, 2 .^ min.(1:2R-1, 2R-1:-1:1), color="gray", linewidth=0.5)
axs[2, 2].semilogy(1:2R-1, TCI.linkdims(qttK2a.tt))
axs[2, 2].semilogy(1:2R-1, TCI.linkdims(qttK2a2.tt))
axs[2, 2].set_xlabel(L"\ell")
axs[2, 2].set_ylabel(L"D_\ell")
axs[2, 2].set_ylim(1.5, 300)
axs[2, 2].set_xticks([1, 10, 19])

tight_layout()

fig.savefig("K2aqtt.pdf")
```
